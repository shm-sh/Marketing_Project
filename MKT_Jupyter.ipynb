```python
q1 = mkt_df['Income'].quantile(0.25)
q3 = mkt_df['Income'].quantile(0.75)
iqr = Q3 - Q1
iqr
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
outliers = mkt_df[(mkt_df['Income'] < lower_bound) | (mkt_df['Income']>upper_bound)]
outliers
```

```python
sns.histplot(mkt_df['Income'], kde = True, color = 'Brown', bins = 30)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Density')
```

```python
mkt_df = mkt_df[(mkt_df['Income'] > lower_bound) & (mkt_df['Income'] < upper_bound)]
```

```python
mkt_df['Income'] = scaler.inverse_transform(data_imputed)
```

```python
from datetime import datetime
current_year = datetime.now().year
mkt_df['Age'] = current_year - mkt_df['Birth_Year']
```

```python
mkt_df['Days_As_Customer'] = mkt_df['Customer_Registration_Date'].max() - mkt_df['Customer_Registration_Date']
```

```python
mkt_df['Days_As_Customer'] = mkt_df['Days_As_Customer'].astype(str).str.replace(' days','')
```

```python
mkt_df['Days_As_Customer'] = pd.to_numeric(mkt_df['Days_As_Customer'], downcast='integer')
```

```python
mkt_df['Total_Purchases'] = mkt_df['Web_Purchases'] + mkt_df['Catalog_Purchases'] + mkt_df['Store_Purchases']
```

```python
mkt_df.info()
```

```python
mkt_df['Total_Spend'] = mkt_df.filter(like='Expense').sum(axis=1)
```

```python
mkt_df['Total_Accepted_Campaign'] = mkt_df.filter(like='Accepted_Campaign').sum(axis=1)
```
